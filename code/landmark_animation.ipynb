```python
def update(frame):
    flipped_frame = frame.copy()  # 원본 데이터 유지
    flipped_frame[:, 1] = 1000 - frame[:, 1]  # y축 값 반전 (0~1000 범위 기준)
    sc.set_offsets(flipped_frame)  # 반전된 좌표 사용
    return sc,
```

```python
# 애니메이션 생성
fig, ax = plt.subplots()
sc = ax.scatter([], [], s=10)
ax.set_xlim(700, 1200)  # x축 범위 (데이터에 맞게 조정)
ax.set_ylim(0, 1000)  # y축 범위 (데이터에 맞게 조정)
ax.set_title("Landmark Animation Pred")

ani = FuncAnimation(fig, update, frames=landmark_preds1[0, :, :, :], interval=50, blit=True)

video = HTML(ani.to_html5_video())
plt.close(fig)  # 중복 출력 방지
video
```
